Menggabungkan data train_features dengan test_festures

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train_features = pd.read_csv('train_features.csv')
test_features = pd.read_csv('test_features.csv')

data = pd.concat([train_features, test_features], axis=1)
```

```python
data.head()
```

Analisis Perbedaan

```python
data.describe() #Melakukan analisis statistik deskriptif
```

Dataset ini memiliki variasi yang cukup besar dalam beberapa kategori seperti belanja makanan dan pengeluaran. Namun, data juga mengindikasikan adanya keseragaman dalam beberapa kategori lainnya, seperti diskon yang digunakan dan jumlah anak.

```python
data.isnull().sum() / len(data) * 100
```

```python
num_cols = [
    'tahun_kelahiran',
    'pendapatan',
    'belanja_buah',
    'belanja_daging',
    'terakhir_belanja',
    'pembelian_toko',
    'pembelian_web',
]
```

```python
corr = data[num_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='viridis', fmt=".2f")
plt.title('Colleration Heatmap')
plt.show()
```

Pada heatmap diatas menunjukan korelasi antar variabel pada dataset test_features dan train_features. Nilai pada heatmap berkisar -1 hingga 1.  Warna kuning (1) menunjukan korelasi sempurna dengan dirinya sendiri, Warna Hijau menunjukan korelasi positif (sekitar 0.3 - 0.7), warna ungu menunjukan korelasi negatif, warna lebih gelap menunjukan korelasi mendekati 0(netral). Pada heatmap diatas menunjukan banyak perbedaan dalam nilai korelasi antar variabel, maka bisa disimpulkan bahwa kedua dataset tersebut memiliki perbedaan karakteristik yang signifikan.
