```python
#Function to count outliners using IQR
def count_outliners_iqr(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 *IQR
    upper_bound = Q3 + 1.5 *IQR
    return ((df < lower_bound) | (df > upper_bound)).sum()

#Count outliners in each numerial column
outliners_count = {}
for col in df.select_dtypes(include=['int64', 'float64']).columns:
    outliners_count[col] = count_outliners_iqr(df[col])
    
#convert the result into a data frame for easier viewing
outliners_count_df = pd.DataFrame(list(outliners_count.items()), columns=['Column', 'Outliner Count'])

#Display the outliner counts DataFrame
outliners_count_df
```

```python
def plot_histogram(df):
    df.hist(bins=10, figsize=(15, 10), layout=(3, 3), color='skyblue', edgecolor='black')
    plt.suptitle('Histogram of Pima Indian Diabetes Dataset Features', fontsize=16)
    plt.show()

plot_histogram(df)
```

```python
# visualisasi boxplot
def plot_boxplots(data):
  plt.figure(figsize=(15,10))
  for i, column in enumerate(data.columns[:-1]):
    plt.subplot(3, 3, i+1)
    sns. boxplot(x='Outcome', y=column, data=data)
    plt.title(f'Box Plot of {column} by Diabetes Outcome')
  plt.tight_layout()
  plt.show()

plot_boxplots(df)
```
